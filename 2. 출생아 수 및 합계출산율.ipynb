```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family']='Malgun Gothic' #Windows
matplotlib.rcParams['font.size']=15 #글자크기
matplotlib.rcParams['axes.unicode_minus']=False #한글 폰트를 사용시, 마이너스 글자 깨지지 않도록함
```

```python
plt.plot(df.index,df['출생아 수']) #df.index는 2012~2020 column을 나타냄.
plt.plot(df.index,df['합계 출산율'])
```

```python
fig,ax1=plt.subplots(figsize=(10,7)) #subplot은 아무표시없으니까 1개의 figure만 출력
ax1.plot(df.index,df['출생아 수'],color='#ff812d')

ax2=ax1.twinx() #x축을 공유하는 쌍둥이 axis
3ax2.plot(df.index,df['합계 출산율'],color='#ffd100')
```

```python
#출생아수 그래프 수정

fig,ax1=plt.subplots(figsize=(13,5)) #subplot은 아무표시없으니까 1개의 figure만 출력
ax1.set_ylabel('출생아 수(천 명)')
ax1.set_ylim(250,700) #출생아 수 y축 250~700로 제한
ax1.set_yticks([300,400,500,600]) #출생아 수 y축값 300,400,500,600
bar=ax1.bar(df.index,df['출생아 수'],color='#ff812d') #출생아수는 막대그래프
for idx,val in enumerate(df['출생아 수']):
    ax1.text(idx,val+2,val,ha='center') #ha= horizontal allignment

ax2=ax1.twinx() #x축을 공유하는 쌍둥이 axis
ax2.set_ylabel('합계 출산율 (가임여성 1명당 명)')
ax2.set_ylim(0,1.5)  #합계출산율 y축 0~1.5로 제한
ax2.set_yticks([0,1])  #합계출산율 수 y축값 0,1
ax2.plot(df.index,df['합계 출산율'],color='#ffd100') #합계 출산율은 꺽은선 그래프
```

```python
#합계 출산율 그래프 수정

fig,ax1=plt.subplots(figsize=(13,5)) #subplot은 아무표시없으니까 1개의 figure만 출력
ax1.set_ylabel('출생아 수(천 명)')
ax1.set_ylim(250,700) #출생아 수 y축 250~700로 제한
ax1.set_yticks([300,400,500,600]) #출생아 수 y축값 300,400,500,600
bar=ax1.bar(df.index,df['출생아 수'],color='#ff812d') #출생아수는 막대그래프
for idx,val in enumerate(df['출생아 수']):
    ax1.text(idx,val+12,val,ha='center') #ha= horizontal allignment

ax2=ax1.twinx() #x축을 공유하는 쌍둥이 axis
ax2.set_ylabel('합계 출산율 (가임여성 1명당 명)')
ax2.set_ylim(0,1.5)  #합계출산율 y축 0~1.5로 제한
ax2.set_yticks([0,1])  #합계출산율 수 y축값 0,1
ax2.plot(df.index,df['합계 출산율'],color='#ffd100',marker='o',markersize=15,linewidth=5,markeredgecolor='r',markeredgewidth=3) #합계 출산율은 꺽은선 그래프
```
